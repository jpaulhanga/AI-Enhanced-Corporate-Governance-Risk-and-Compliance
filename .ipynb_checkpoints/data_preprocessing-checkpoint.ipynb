```python
import pandas as pd 
```

```python
file_path_corporate_finance  = '/home/project/Documents/mgmt/Corporate_Financial_Risk_Assessment_Data.csv'

df_corporate_finance = pd.read_csv(file_path_corporate_finance)
df_corporate_finance.columns = [c.strip().lower() for c in df_corporate_finance.columns]

df_corporate_finance = df_corporate_finance.rename(columns={"company_id":"ticker"})

if "date" in df_corporate_finance.columns:
    df_corporate_finance["year"] = pd.to_datetime(df_corporate_finance["date"], errors="coerce").dt.year
    
CF=df_corporate_finance[["ticker","year"]].dropna().drop_duplicates()
display(CF)
```

```python
file_path_SEC_financial_summary  = '/home/project/Documents/mgmt/SEC_DATA_EDGAR_3 - Sheet1.csv'

df_financial_summary = pd.read_csv(file_path_SEC_financial_summary)
df_financial_summary.columns = [c.strip().lower() for c in df_financial_summary.columns]
df_financial_summary["ticker"] = df_financial_summary["company"].replace({
    "Tesla": "TSLA",
    "Microsoft": "MSFT",
    "Apple": "AAPL"
})
df_financial_summary = df_financial_summary.rename(columns={"fiscal year":"year"})
    
#df_financial_summary.head()
FS=df_financial_summary[["ticker","year"]].dropna().drop_duplicates()
display(FS)
```

```python
file_path_public_company_ESG  = '/home/project/Documents/mgmt/data.csv'

df_public_company_ESG = pd.read_csv(file_path_public_company_ESG)
df_public_company_ESG.columns = [c.strip().lower() for c in df_public_company_ESG.columns]

if "last_processing_date" in df_public_company_ESG.columns:
    df_public_company_ESG["year"] = pd.to_datetime(df_public_company_ESG["last_processing_date"], format="%d-%m-%Y", errors="coerce").dt.year

PC=df_public_company_ESG[["ticker","year"]].dropna().drop_duplicates()
display(PC)
```

```python
file_path_SP_500_ESG = '/home/project/Documents/mgmt/SP 500 ESG Risk Ratings.csv'

df_SP_500_ESG= pd.read_csv(file_path_SP_500_ESG )
df_SP_500_ESG.columns = [c.strip().lower() for c in df_SP_500_ESG.columns]
df_SP_500_ESG = df_SP_500_ESG.rename(columns={"symbol":"ticker"})

SP = df_SP_500_ESG[["ticker"]].dropna().drop_duplicates()
display(SP)
```

```python
#Unified dataset aligned on (ticker, year)

unified = (CF.merge(FS, on=["ticker", "year"], how="outer").merge(PC, on=["ticker", "year"], how="outer").merge(SP, on="ticker", how="outer"))
unified['year'] =unified['year'].astype('Int64')
display(unified)
```

```python
unified.to_csv("unified_dataset.csv", index=False)
print("Saved as 'unified_dataset.csv'")
```
